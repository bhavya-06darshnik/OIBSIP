UNDERSTANDING DATA:
1. SEPAL: These are leaf-like structures enclosing the bud of the flower
Can be seen at the lower part of the flower
It protects and folds over the closed buds from the external surrounding
Usually, a flower has 2-5 sepals.

2. PETAL: These are modified leaves surrounding the reproductive structures of flowers
These are unusually shaped or brightly coloured attracting pollinators.

IN GIVEN DATASET WE WILL FIND RELATION BETWEEN FEATURES SUCH AS; SEPAL (LENGTH & WIDTH) & PETAL (LENGTH & WIDTH) OF DIFFERENT IRIS SPECIES>
